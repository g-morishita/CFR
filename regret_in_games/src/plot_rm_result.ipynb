```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot the difference between the regret matching strategy and the nash equilibrium strategy
num_iterations = 250
rsp_results = np.load("rsp_result.npy")
nash_eq_strategy = np.array([ 1.0 / 3, 1.0 / 3, 1.0 / 3])
diff = np.abs(rsp_results - nash_eq_strategy).sum(axis = 1)
fig, ax = plt.subplots()
ax.plot(range(1, num_iterations), diff[:num_iterations - 1])
ax.set_xlabel("the number of iterations")
ax.set_ylabel("gap between NE and RM")
ax.set_title("Regret Maching Algorithm converges to NE")
```
